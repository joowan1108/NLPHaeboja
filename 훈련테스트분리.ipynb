데이터를 적절하게 분리하여 train set, validation set, test set으로 나누어야 한다.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

zip 함수: 동일한 개수를 가지는 sequence 자료형에서 각 순서에 등장하는 원소들끼리 묶어준다. 즉, 두 sequence가 존재한다면, index가 같은 것들끼리 하나의 데이터로 만드는 것이다. 리스트의 리스트 구성에서 zip를 통해 x와 y를 분리하는데 사용할 수 있다.

```python
#x,y를 분리하는데 사용
x,y = zip(['a', 1], ['b', 2], ['c', 3])
print("x 데이터 : ", x)
print("y 데이터 : ", y)
```

```python
sequences = [['a', '1'], ['b','2'],['c',3]]
x,y = zip(*sequences)
print("x 데이터 : ", x)
print("y 데이터 : ", y)
```
