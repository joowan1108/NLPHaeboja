#선형회귀
데이터에 가장 fit한 y=wx+b를 찾는 것.
즉, 최적의 w와 b를 찾는 것이다.

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(1)
```

#manual_seed() 하는 이유
다른 컴퓨터에서도 random seed를 고정시킨다.
이로 인해 동일한 값이 나온

```python
print(x_train)
print(x_train.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
W = torch.zeros(1,requires_grad=True)
#크기가 1인 tensor. 값은 0
#requires_grad=True -> 학습을 통해 계속 변하는 값임을 명시하는 방법
print(W)
```

```python
b = torch.zeros(1,requires_grad=True)
print(b)
```

현재는 y=0*x+0 이다.

#가설 세우기
코드상으로 직선의 방정식에 해당하는 가설을 선언한다. 가설의 형태를 정의한다.

```python
hypo = x_train*W + b
print(hypo)
```

#비용 함수 Cost Function 선언하기
여기서는 Mean Squared Error을 사용

```python
cost = torch.mean((hypo - y_train)**2)
print(cost)
```

#최적화 방법 정의
경사 하강법을 사용하고자 한다

```python
optimizer = optim.SGD([W,b], lr=0.01)
#학습 대상 W,b를 설정하고 SGD를 사용하고 learning rate를 0.01로 설정하였다
```

```python
#optimizer의 초기 미분값을 0으로 설정해야 한다. 이렇게 초기값을 설정해야 다음 기울기를 구할 수 있다
optimizer.zero_grad()

#비용 함수를 미분하여 gradient 계산
cost.backward()
#requires_grad=True가 설정된 모든 값을 업데이트할 때 필요한 미분값을 계산한다.

#W와 b를 실제로 업데이트
optimizer.step()
```

```python
epochs = 1999
for e in range(epochs+1):

  hypo = x_train*W+b
  cost = torch.mean((hypo-y_train)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if e%100 ==0:
    print("Epoch {:4d}/{} W:{:.3f}, b:{:.3f} Cost: {:.6f}".format(e, epochs, W.item(), b.item(), cost.item()))
```
