#토큰화

얻은 corpus 데이터가 필요에 맞게 전처리되지 않은 상태라면?

토큰화, 정제, 정규화를 진행해야 한다. corpus에서 토큰이라고 불리는 단위로 나누는 작업을 토큰화라고 한다. 보통 의미있는 단위로 corpus를 나눈다.



#단어 토큰화

띄어쓰기, 특수 문자, 구두점 등을 모두 제거하면 언어에 따라 그 단어가 의미를 잃을 수 있다. 또는 그런 특징에 의해 토큰화가 애매해질 수 있다.

ex) Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop.

에서 Don't, Jone's 의 '를 어떻게 분리할 것인가?

nltk는 word_tokenize, WordPunctTokenizer 등의 도구를 제공하는데 각각 다르다.

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

```python
print("word_tokenize의 결과")
print(word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

Don't -> Do, n't로 분리

Jone's -> Jone, 's로 분리

```python
print('WordPunctTokenizer')
print(WordPunctTokenizer().tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

Don't -> Don, ', t

Jone's -> Jone, ', s로 분리

```python
print("keras text_to_word_sequences")
print(text_to_word_sequence("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

#토큰화할 때 character가 아닌 것을 다 제거하면 안되는 이유



1. 구두점이나 특수 문자를 단순 제외해서는 안 된다.

구두점이나 특수 문자 또한 의미를 가지거나 다른 기능을 할 수 있다. 예를 들어, 마침표는 문장을 구분해주는 역할을 한다. 또는, Ph.D, AT&T 같은 경우, 12,000 의 , 경우, 다 의미가 존재한다.

2. 줄임말과 단어 내에 띄어쓰기가 있는 경우

What're 등, '은 단어가 압축되어있다는 의미로 사용된다, 이때 re를 clitic이라고 하는데, 단어를 줄임말로 사용할 때 생기는 형태를 말한다.



#표준 토큰화 예제

표준으로 사용되는 토큰화 (Penn Treebank Tokenization) 규칙

1. '-' hyphen 으로 구성된 단어는 하나로 유지한다

2. doesn't와 같이 '로 clitic가 같이 존재하는 단어는 분리해준다.



```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

text = "Starting a home-based restaurant may be an ideal. it doesn't have a food chain or restaurant of their own."

print(tokenizer.tokenize(text))
```

#문장 토큰화

문장 단위로 작업을 하고싶은데 corpus가 문장 단위로 구분되어 있지 않을 떄 사용한다.

. ? ! 등으로 구분하고 싶어도 마침표가 다른 의미를 지닐 수 있기 때문에 그걸 기준으로 하면 안된다.

예시) IP.192.168.56.31 서버 정보를 aaa@gmail.com으로 보내줘 같은 것



nltk에서는 sent_tokenize를 지원하여 영어 문장의 토큰화를 수행한다.

```python
from nltk.tokenize import sent_tokenize

text = "His barber kept his word. But keeping such a huge secret to himself was driving him crazy. Finally, the barber went up a mountain and almost to the edge of a cliff. He dug a hole in the midst of some reeds. He looked about, to make sure no one was near."
print('문장 토큰화1 :',sent_tokenize(text))
```

Ph.D 같이 .이 특정 의미를 지니고 있을 때도 분리할 수 있을까?

```python
text = "I am actively looking for Ph.D. students. and you are a Ph.D student."
print('문장 토큰화2 :',sent_tokenize(text))
```

#한국어에서 토큰화가 어려운 이유

#1. 교착어이기 때문이다.
교착어란 조사, 어미 등을 붙여쓰는 언어이다. 조사, 어미들은 붙어있는 단어와 다른 의미를 지니는데 붙어있기 때문에 띄어쓰기를 기준으로 분리하면 기존의 단어와 조사/어미가 하나의 단어가 된다.

따라서, 형태소의 개념이 중요하다. 형태소란 뜻을 가진 가장 작은 말의 단위이다.

자립 형태소: 접사 어미 조사와 상관없이 자립하여 사용할 수 있는 형태소. 체언(명사, 대명사, 수사), 수식언(관형사, 부사), 감탄사

의존 형태소: 다른 형태소와 결합하여 사용되는 형태소. 접사, 어미, 조사, 어간

문장: 에디가 책을 읽었다

띄어쓰기를 기준으로 토큰화를 진행한다면,['에디가', '책을', '읽었다']가 된다.

하지만 형태소를 기준으로 토큰화를 진행한다면

자립 형태소 : 에디, 책
의존 형태소 : -가, -을, 읽-, -었, -다


따라서, 한국어는 어절 단위가 아니라 형태소 토큰화를 수행해야 한다.

#2. 띄어쓰기가 잘 지켜지지 않는다.
한국어는 띄어쓰기를 하지 않아도 이해할 수 있는 특징을 지닌다. (모아쓰기)

예) 제가이렇게띄어쓰기를전혀하지않고글을썼다고하더라도글을이해할수있습니다.

하지만 영어는 띄어쓰기를 하지 않으면 글자들이 융합되기 때문에 이해하기가 어렵다. (풀어쓰기)

예) Tobeornottobethatisthequestion





또 다른 어려운 점은 단어가 표기는 같아도 품사에 따라 의미가 크게 바뀐다는 것이다. 못은 nail 이 되면서 can't으로 해석될 수도 있다.

따라서, 완벽하게 이해하기 위해서는 각 단어가 어떤 품사로 사용되었는지를 구분해야 한다.

이를 품사 태깅, part of speech tagging이라고 한다.

#품사 태깅

nltk는 Penn Treebank POS Tags를 기준으로 품사 태깅을 할 수 있다.



```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
text = "I am actively looking for Ph.D. students. and you are a Ph.D. student."
tokenized_sentence = word_tokenize(text)
print('단어 토큰화 :',tokenized_sentence)
print('품사 태깅 :',pos_tag(tokenized_sentence))
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma

okt = Okt()
kkma = Kkma()
```

```python
print('OKT 형태소 분석 :',okt.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 품사 태깅 :',okt.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 명사 추출 :',okt.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

```python
print('꼬꼬마 형태소 분석 :',kkma.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 품사 태깅 :',kkma.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 명사 추출 :',kkma.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```
