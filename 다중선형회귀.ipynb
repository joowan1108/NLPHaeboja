#다중 선형 회귀

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
torch.manual_seed(1)
```

```python
# 훈련 데이터
x1_train = torch.FloatTensor([[73], [93], [89], [96], [73]])
x2_train = torch.FloatTensor([[80], [88], [91], [98], [66]])
x3_train = torch.FloatTensor([[75], [93], [90], [100], [70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
# 가중치 w와 편향 b 초기화
w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
#optimizer 설정
optimizer = optim.SGD([w1, w2, w3, b], lr=1e-5)

epochs = 2000
for e in range(epochs+1):
  #hypothesis
  hypo = x1_train*w1 + x2_train*w2 + x3_train*w3 + b
  #cost
  cost = torch.mean((hypo-y_train)**2)

  #cost 값을 기반으로 최적화 과정 진행
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if e%100==0:
    print('Epoch {:4d}/{} w1: {:.3f} w2: {:.3f} w3: {:.3f} b: {:.3f} Cost: {:.6f}'.format(
            e, epochs, w1.item(), w2.item(), w3.item(), b.item(), cost.item()
        ))
```

#벡터 연산으로 바꾸어보기

```python
x_train  =  torch.FloatTensor([[73,  80,  75],
                               [93,  88,  93],
                               [89,  91,  80],
                               [96,  98,  100],
                               [73,  66,  70]])
y_train  =  torch.FloatTensor([[152],  [185],  [180],  [196],  [142]])
print(x_train.shape)
print(y_train.shape)
```

```python
# 가중치와 편향 선언
W = torch.zeros((3, 1), requires_grad=True) #3x1 크기로 설정하기
b = torch.zeros(1, requires_grad=True)
```

```python
#optimizer 설정
optimizer = optim.SGD([W,b], lr=1e-5)
epochs = 2000
for e in range(epochs+1):
  #hypothesis 선언
  hypo = x_train.matmul(W)+b

  #cost 계산
  cost = torch.mean((hypo-y_train)**2)

  #cost기반으로 최적화
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if e%100==0:
    print('Epoch {:4d}/{} hypothesis: {} Cost: {:.6f}'.format(
        e, epochs, hypo.squeeze().detach(), cost.item()
    ))


```
